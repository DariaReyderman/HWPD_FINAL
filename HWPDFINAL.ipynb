```python
import numpy as np 
import pandas as pd
from IPython.display import display

import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(6.4, 4.8))
axes1 = fig.add_axes([0, 0, 1, 1])
x = np.arange(0, 5, 0.1)
y = np.sin(x)
axes1.set_title("Sine function")
axes1.set_xlabel("X")
axes1.set_ylabel("Sin(X)")
axes1.plot(x, y)

axes2 = fig.add_axes([1.2, 0, 1, 1])
x = np.arange(0, 5, 0.1)
y = np.cos(x)
axes2.set_title("Cosine function")
axes2.set_xlabel("X")
axes2.set_ylabel("Cos(X)")
axes2.plot(x, y)
plt.show()
```

### Exercise 2. Subplot

```python
# Data
months = ["Jan", "Feb", "Mar", "Apr", "May"]
temperature = [12, 14, 17, 21, 25]
rain = [80, 60, 55, 40, 20]

# Subplots
fig, ax = plt.subplots(1, 2, figsize=(6, 6))

ax[0].plot(months, temperature, label="Temterature", color='r')
ax[0].set_title("Temperature per Months")
ax[0].grid(True)
ax[0].legend()

ax[1].bar(months, rain, label="Rain", color='b')
ax[1].set_title("Rain per Months")
ax[1].grid(True)
ax[1].legend()

fig.suptitle('The Weather Statistics', fontsize=16)

plt.tight_layout()
plt.show()
```

### Exercise 3. Histogram

```python
grades = [70, 85, 90, 76, 65, 95, 88, 73, 80, 78, 67, 92, 60, 83, 89, 55, 91, 74, 69, 87, 81, 79, 62, 77, 93]
fig = plt.figure()
ax = fig.add_axes([0, 0, 0.7, 1])
ax.hist(grades, bins=10, color='#5e32a8', alpha=0.7)
ax.set_title("Exam Grades")
ax.set_xlabel("Grades")
ax.set_ylabel("Frequency")
plt.show()
```

### Cleaning up missing information

```python
students = pd.read_csv("students.csv")
teachers = pd.read_csv("teachers.csv")
courses = pd.read_csv("courses.csv")
student_course = pd.read_csv("student_course.csv")
```

```python
display(students.dropna().drop_duplicates())
display(teachers.dropna().drop_duplicates())
display(courses.dropna().drop_duplicates())
display(student_course.dropna().drop_duplicates())
```

### Exercise 1

```python
gender_counts = students['gender'].value_counts()
fig = plt.figure()
ax = fig.add_axes([0, 0, 0.5, 0.5])
ax.bar(gender_counts.index, gender_counts.values, color=['#1307f2','#f2076d'])
ax.set_title('Male vs Female')
ax.set_xlabel('Gender')
ax.set_ylabel('Count')
plt.grid(True)
plt.show()
```

### Exercise 2

```python
age = students['age'].sort_values(ascending=True)
fig = plt.figure()
ax = fig.add_axes([0, 0, 0.8, 0.8])
ax.hist(age, bins=20, color='#def01a', alpha=0.7)
ax.set_title("Students' Age")
ax.set_xlabel("Age")
ax.set_ylabel("Count")
plt.show()
```
